# Premier-League-Club-Investment-Analysis

Context:
Company ABC have invested in top-tier sports teams. The dataset in their possession comprises crucial information about all the teams that have participated in premier league (assume that it is the data for all teams). It includes data on the number of goals scored and conceded by each team, the number of times they have finished in the top two positions and other relevant details.

Data: 
Premier League Final Data.csv- : The data set contains information on all the teams so far participated in all the premier league tournaments.
    
Data Dictionary:
 - Club: Name of the football club
 - Matches: Number of matches the club has played in the Premier League
 - Wins: Number of matches won by the club in the Premier League
 - Loss: Number of matches lost by the club in the Premier League
 - Draws: Number of matches drawn by the club in the Premier League
 - Clean Sheets: Number of matches in which the club has prevented the opposing side from scoring
 - Team Launch: Year in which the club was founded
 - Winners: Number of times the club has won the Premier League
 - Runners-up: Number of times the club has finished as runners-up in the Premier League

Primary Objective:
The management of Company ABC aims to invest in some of the top-performing clubs in the English Premier League. To aid in their decision-making process, the analytics department has been tasked with creating a comprehensive report on the performance of various clubs. However, some of the more established clubs have already been owned by the competitors. As a result, Company ABC wishes to identify the clubs they can approach and potentially invest to ensure a successful and profitable deal.

# SECTION 2- CLEANING THE DATA

```python
#removing the numeric values attached to each of the club name

df['Club']=df['Club'].str.replace('\d+','')

#\d+ means one or more digits at the starting of any string.
#str.replace replace a given pattern with the desired pattern.
```

```python
df.head()
```

The numeric values have been removed before the club names.

```python
df['Winners'].isnull().any()
```

```python
df['Winners'].value_counts()
```

```python
#removing null values in winners column

df['Winners'].fillna(0,inplace=True)

#fill na used to fill null values and null values will be replaced by 0.
#inplace=true means that the original dataset will get updated when this query will be executed.
```

```python
df
```

Now we can see that the null values in the Winners column have been replaced by 0.0. Therefore, this column can now be analysed.

```python
#removing null values of runners-up column

df['Runners-up'].fillna(0,inplace=True)
```

```python
df
```

The null values of Runners-up column have been replaced with 0 but still there are '-' these symbols present in the column.

```python
df['Runners-up']=df['Runners-up'].str.replace('-','0')
```

```python
df
```

```python
df['Runners-up'].fillna(0,inplace=True)

#for direct replace of - to 0 we can use .replace instead of str.replace in the above step.
```

```python
df
```

So now the null values of Runners-up column have been replaced with 0. Now object type should be changed to int type of runners up column.

```python
df.info()
```

```python
#changing object type to int type

df['Runners-up']=pd.to_numeric(df['Runners-up'],errors='coerce')
df['Runners-up']=df['Runners-up'].astype('Int64')
```

```python
df.info()
```

So now the object data type is changed to int data type. So now it can be used for analysis.

```python
df['TeamLaunch']
```

```python
#now correcting the date type

#converting to datetime format
df['TeamLaunch']=pd.to_datetime(df['TeamLaunch'],errors='coerce')
#converting to YYYY format
df['TeamLaunch']=df['TeamLaunch'].dt.strftime('%Y')
```

```python
df
```

So now we have converted the time launch column to YYYY format.

```python
df.info()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.hist(df['Matches Played'])
plt.xlabel('Number of matches played')
plt.ylabel('Frequency')
plt.title('Histogram showing number of matches played')
```

Majority of the teams have played matches less than 400 and only few team are there who have played more than 900 matches.

```python
#teams which played more than 900 matches

df[df['Matches Played']>=900]['Club']
```

There are 11 teams which have high experience. Company ABC want clubs which may win in future and can even have less experience. So we will drop these 11 clubs and will consider only the remaining clubs for further analysis.

```python
df=df[df['Matches Played']<900]
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
df['Win_probability']=df['Win']/df['Matches Played']
```

```python
club_with_highest_win_probability=df.loc[df['Win_probability'].idxmax(),'Club']
```

```python
df_top5=df.sort_values(by='Win_probability',ascending=False).head(5)
plt.figure(figsize=(10,6))
plt.bar(df_top5['Club'],df_top5['Win_probability'],color='skyblue')
plt.show()
```

```python
club_with_highest_win_probability
```
